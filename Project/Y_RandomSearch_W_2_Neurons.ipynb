```python
import numpy
import keras
import pandas as pd
from sklearn.model_selection import RandomizedSearchCV
from keras.models import Sequential
from keras.layers import Dense
import warnings
warnings.filterwarnings('ignore')
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
data = pd.read_csv('/content/drive/MyDrive/dataset-org.csv')
```

```python
data.head()
```

```python
X = data.iloc[:,0:178]
Y = data.iloc[:,178]
```

```python
def create_my_model(optimizer='adam', activationL='relu', neurons=0.01):
  mymodel=Sequential()
  mymodel.add(Dense(16,input_dim=178,activation='relu'))
  mymodel.add(Dense(1,activation='sigmoid'))

  mymodel.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
  return mymodel
```

```python
model=KerasClassifier(build_fn=create_my_model)
```

```python
batchSize=[10,20,40,60,80,100]
epochs=[10,30,50]
optimizer=['SGD', 'Adadelta', 'RMSprop', 'Adagrad', 'Adam']
activationL= ['relu', 'sigmoid', 'softplus', 'softsign', 'tanh', 'selu', 'elu', 'exponential', 'LeakyReLU']
neurons= (10,100)
```

```python
parameter_rdm=dict(batch_size=batchSize, epochs=epochs, optimizer=optimizer, activationL=activationL, neurons=neurons)

rdm_search =RandomizedSearchCV(estimator=model,param_distributions= parameter_rdm, n_jobs=-1,cv=3)
rdm_result =rdm_search.fit(X,Y)
```

```python
print("Best:%f using %s"%(rdm_result.best_score_,rdm_result.best_params_))
```
